### 2번 차원축소 (1번 이어서)
#### 2-1 차원축소 방법 2가지 이상 비교하고 한 가지 선택

차원 축소는 차원이 늘어남에 따라 데이터 간의 거리가 멀어지는(데이터가 전체 영역에서 설명할 수 있는 부분이 줄어드는) 현상인 "차원의 저주" 문제를 해결하기 위한 방법론이다. 또한, 분석이 어려운 고차원의 데이터를 차원 축소로 인해 계산 비용을 감소하고 노이즈를 감소하는 효과를 가져온다. 이러한 차원 축소의 방법론에는 주성분 분석(pca), LSA, 요인 분석(FA), 독립성분 분석(ICA), t-SNE 등이 있다.

- 주성분 분석 : 주성분 분석은 데이터에서 가장 많은 정보를 담고 있는 방향으로 변수를 축소하는 방법이다. 고차원의 데이터 손시를 최소화하면서 효율적으로 축소할 수 있는 방법론인데 반해, 데이터를 선형적으로 축소하기 때문에 비선형성을 반영하지 못한다.

- LSA : LSA는 절단된 특이값 분해 방식을 사용하는 차원 축소 방법이다. PCA와 달리 데이터를 중앙에 배치하지 않고 진행한다.

- 요인 분석 : 요인 분석은 다수 변수들 간의 상관관계를 분석하여 공통 차원들을 통해 축약하는 통계적인 기법이다. 다수의 변수들을 정보 손실을 최소화하면서 소수의 요인으로 축약하는 방법이다. 요인분석은 관측되지 않는 잠재 변수에 대하여 이런 연결된 분산을 찾는다.

- 독립성분 분석 : 독립성분 분석은 다변량의 신호를 통계적으로 독립적인 하부 성분으로 분리하는 계산 방법이다.

- t-SNE : t-SNE는 데이터 포인트 간의 유사성을 결합 확률로 변환하고 저차원 임베딩과 고차원 데이터의 결합 확률 사이의 Kullback-Leibler 발산을 최소화하는 분포에 따라 차원을 축소하는 방법이다.

#### 2-2 추천한 한 가지를 실제로 수행하고 선택한 이유 설명

이 중에서 지금 다루고 있는 상관관계가 높은 데이터에서의 차원 축소 기법으로는 pca나 요인 분석이 적절하다고 볼 수 있다. 특히, 요인 분석은 통계적으로 상관관계를 분석하여 공통 차원들을 통해 축약하는 방법론이므로 이 데이터에 적합하기 때문에 요인 분석으로 차원축소를 진행한다.

```python
from sklearn.decomposition import FactorAnalysis
import matplotlib.pyplot as plt

best_dim = 5

fa_model = FactorAnalysis(n_components = best_dim)
fa_model.fit(df1[features])
FA_result = fa_model.transform(df1[features])

max_var = pd.DataFrame(fa_model.components_, index = [f"latent{i}" for i in range(1, best_dim+1)], columns=features)

fig, axes = plt.subplots(best_dim, 1, figsize=(8, 25))
for i, ax in enumerate(axes):
    max_var.iloc[i, :].plot.bar(ax=ax, rot=90)
    ax.set_xticks(list(range(len(features))))
    ax.set_xticklabels([" " for i in range(len(features))])
    twin_ax = ax.twinx()
    twin_ax.set_ylabel(max_var.index[i], fontsize=15)
    ax.axhline(y=0, linestyle='dashed', color='red')
    ax.set_xticks(list(range(len(features))))
    ax.set_xticklabels(list(features))
axes[best_dim//2].set_ylabel("Component Loading (maximum variance)", fontsize=20)
plt.show()
```

### 3 오버샘플링 / 언더샘플링 (1번 이어서)
#### 3-1 오버샘플링과 언더샘플링의 장단점 비교 및 선택 구현

오버샘플링과 언더샘플링은 클래스의 불균형을 해소하기 위한 기법이다. 오버 샘플링은 부족한 데이터를 가진 클래스의 데이터를 RandomOverSampler, SMOTE 등의 기법을 이용하여 데이터를 생성하는 방법이고, 언더 샘플링은 RandomUnderSampler, ClusterCentroids 등의 기법을 이용하여 충분한 데이터를 가진 클래스의 데이터를 제거하는 방법이다.

오버 샘플링의 장점은 데이터를 생성하는 방법론이기 때문에 데이터의 양이 중요한 머신러닝에 좀더 적합한 방법론이라고 볼 수 있다. 그러나 생성된 데이터가 그 클래스를 대표하는 feature를 가지고 있지 않을 가능성이 높고 generalize 성능이 떨어지는 단점이 존재한다.

언더 샘플링의 장점은 데이터가 적은 클래스에 맞게 데이터를 제거하므로 실제 관측된 데이터로만 분석이 가능하기 때문에 좀더 generalize 성능이 좋을 수 있다. 하지만 이는 데이터의 양이 충분할 경우에만 가능하기 때문에 현실에서 적용하기 어려운 측면이 있다.

현재 다루고있는 데이터의 양이 1193개로 매우 적은 편에 속하므로 오버샘플링이 좀더 적합하다고 볼 수 있다.

```python
import warnings
from time import time
from sklearn.model_selection import learning_curve
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
warnings.filterwarnings('ignore')

xgboost = XGBClassifier(random_state=42)
logistic_regression = LogisticRegression(random_state=42)

for df_name in ['df1_ROS', 'df1_SM']:
    for model_name in ['logistic_regression', 'xgboost']:
        print('='*20,'sampling :', df_name, '='*20)
        print('='*20 , 'model :', model_name, '='*20)
        df = globals()[df_name]
        model = globals()[model_name]

        metric = 'accuracy'
        train_sizes, train_scores, test_scores = learning_curve(
            estimator=model,
            X=df[features],
            y=df[target],
            cv=5,
            scoring=metric,
            train_sizes = [1, len(df) // 5, len(df) // 5 * 2, len(df) // 5 * 3,  len(df) // 5 * 4]
        )

        plt.subplots(figsize=(10,8))
        plt.plot(train_sizes, train_scores.mean(axis=1), label="train")
        plt.plot(train_sizes, test_scores.mean(axis=1), label="validation")

        plt.title("Learning Curve")
        plt.xlabel("Training Set Size")
        plt.ylabel(metric)
        plt.legend(loc="best")

        plt.show()
        
        x_train, x_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.1, random_state=42)
        model.fit(x_train, y_train)
        
        start = time()
        y_pred = model.predict(x_test)
        end = time()

        # 정확도 평가
        cr = classification_report(y_test, y_pred)
        print()
        print(cr)
        print('inference speed:', end - start)
        print()
```

### 4 이상탐지 (1번 이어서)
#### 4-1 이상탐지 모델 2가지 이상 기술, 장/단점 설명

이상탐지는 어떤 모델을 사용하여 이상치를 탐지하여 처리하는 과정을 말한다. 예를 들어, 스팸분류 모델일 경우에 스팸인 이메일의 특징을 추출하여 해당 이메일이 스팸이라면 0, 아니라면 1 등의 분류 형태로 이상치를 탐지하게 된다. 이러한 이상탐지 모델로는 나이브 베이즈 분류, Kmeans 등이 있다.

- 나이브 베이즈 분류 : 나이브 베이즈 분류는 특성들 사이의 독립을 가정하는 베이즈 정리를 활용한 확률 분류기로서, 조건부 확률을 이용해 새로 들어온 데이터가 기존의 데이터와 비슷한 특징을 가지고 있는지 판단하여 분류를 하는 모델이다. 이러한 모델은 훈련 과정을 이해하기 쉽고 빠르게 이상치를 분류할 수 있지만, 각 특징들을 단순하게 취급하여 성능이 떨어지는 측면도 존재한다.
- Kmeans : Kmeans는 군집화 알고리즘으로서 각 변수들의 특징들간의 거리를 측정하여 가까운 거리의 변수들을 하나의 군집으로 묶는 알고리즘을 말한다. 이러한 알고리즘은 어떠한 특징들로 구성된 군집이 정의되고 이상치라고 정의한 데이터가 모여있는 군집으로 구별하거나 혹은 정상 군집에서 멀리 떨어진 데이터가 있을 경우 이러한 데이터를 이상치로 판단하는 모델이다. 이러한 모델은 어떠한 라벨이 필요가 없으므로 대규모 데이터에서 분석이 가능하다는 장점이 있지만, 분석가가 군집 갯수 등 각 군집의 특성을 파악해야 한다는 단점이 있다.

#### 6번
12건의 광고시간 측정. 평균 15.5초, 분산 3.2초 일때 90% 신뢰구간은?

```python
import numpy as np
from scipy.stats import t

x = 15.5
n = 12
s = 3.2

conf_a = 0.1
df = n-1

SE = s / np.sqrt(n)
conf_t = t.ppf(1-conf_a / 2, df)
ME = conf_t * SE

print('90% 신뢰구간 :', x - ME, '~', x + ME)
```

#### 7-1 귀무가설과 대립가설을 세우시오

- H0 : 강의 상류와 하류의 생물 다양성 점수는 0이다(차이가 존재하지 않는다).

#### 7.2 검정 통계량과 유의 확률을 구하고, 연구가설 채택여부를 검정하시오.

```python
# 종속성을 제거하기 위해 up과 down의 차이를 구한 후(대응표본) 검정 진행
from scipy.stats import ttest_rel

tstat, p = ttest_rel(df2['up'], df2['down'], alternative='two-sided')
print('검정 통계량 :', tstat)
print('유의확률 :', p)
if p < 0.05:
    print('유의수준 5% 하에서 귀무가설 기각, 대립가설 채택')
else:
    print('유의수준 5% 하에서 귀무가설 채택')
```

#### 8-2 
`8-1` 모델의 회귀계수를 활용하여 temperature: 10.5, wind: 8.2, precipitation: 3.5 일때 `user_counts`를 예측하시오

지하철 호선과 월별, 승객 수 간 상관관계가 있는지 확인(Type III Anova 사용)

- 9.1 귀무가설과 대립가설 설정하라

- 9.2 통계량을 구하고, 대립가설 채택여부를 기술하라

```python
import pandas as pd

df4 = pd.read_csv('/Users/yuseonjong/ADP_study/10주차_0730/data/subway.csv')
df4
```

- H0 : 지하철 호선과 월, 승객 수 간의 상관관계가 존재하지 않는다.
- H1 : 지하철 호선과 월, 승객 수 간의 상관관계가 존재한다.

```python
# 1. 정규성 검정(Levene)
# H0 : 표본 모집단이 정규성을 만족한다.
# H1 : 표본 모집단이 정규성을 만족하지 않는다.

from scipy.stats import shapiro

for c in df4.columns:
    sstat, p = shapiro(df4[c])
    print('='*3,c,'='*3)
    print('p-value', p)
    if p < 0.05:
        print('유의확률 5% 하에서 귀무가설 기각, 대립가설 채택')
        print('정규성 만족하지 않음')
    else:
        print('유의확률 5% 하에서 귀무가설 채택')
        print('정규성 만족')
    print()
```

```python
# 2. 등분산성 검정
# H0 : 모든 그룹들의 모분산은 동일하다.
# H1 : 그룹들 적어도 한 그룹의 모분산이 동일하지 않다.

from scipy.stats import levene

lstat, p = levene(df4['line'], df4['month'], df4['total'], center='mean')
print('검정 통계량', lstat)
print('p-value', p)

if p < 0.05:
    print('유의확률 5% 하에서 귀무가설 기각, 적어도 한 그룹의 모분산이 동일하지 않음')
else:
    print('유의확률 5% 하에서 귀무가설 채택, 등분산')
```

```python
# 비정규성이므로 kruskal 검정 시행

from scipy.stats import kruskal

kstat, p = kruskal(df4['line'], df4['month'], df4['total'])
if p < 0.05:
    print('유의확률 5% 하에서 귀무가설 기각, 대립가설 채택')
else:
    print('유의확률 5% 하에서 귀무가설 채택')
```

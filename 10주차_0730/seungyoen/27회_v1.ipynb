```python
def scatterPlot(xDF, yDF, algoName):
    tempDF = pd.DataFrame(data=xDF.loc[:,0:1])
    tempDF = pd.concat((tempDF, yDF), axis=1, join="inner")
    tempDF.columns = ['First Vector', 'Second Vector', 'Label']
    sns.lmplot(x="First Vector", y="Second Vector", hue="Label", data=tempDF, fit_reg=False)
    ax = plt.gca()
    ax.set_title("Seperation of Observations using " + algoName)
```

- RandomOverSampler, SMOTE(Borderline, Kmeans, SVM), ADASYN 등

- RandomUnderSampler,ClusterCentroids, CondenseNearestNeighbour, AllKNN등이 있음
#### Pros and Cons
- Undersample: 데이터 제거로 인한 정보 손실 발생

```python
from imblearn import under_sampling, over_sampling
from imblearn.under_sampling import RandomUnderSampler, ClusterCentroids
from imblearn.over_sampling import RandomOverSampler, SMOTE
from collections import Counter
```

#### 6번
12건의 광고시간 측정. 평균 15.5초, 분산 3.2초 일때 90% 신뢰구간은?

```python
from scipy.stats import norm, t
```

```python
n = 12
mean = 15.5
var = 3.2
sd = np.sqrt(var)
```

```python
t_value = t.ppf(.9, n-1)
```

```python
upper = mean + sd / np.sqrt(n) * t_value
lower = mean - sd / np.sqrt(n) * t_value
```

```python
print(lower, upper)
```

```python
t.interval(confidence=.9, loc=mean, scale=np.sqrt(var/n), df=n-1)
```

#### 9번
지하철 호선과 월별, 승객 수 간 상관관계가 있는지 확인(Type III Anova 사용)

- 9.1 귀무가설과 대립가설 설정하라

- 9.2 통계량을 구하고, 대립가설 채택여부를 기술하라

```python
import pandas as pd

df4 = pd.read_csv('data/subway.csv')
df4
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
# H0: 지하철 호선과 월별, 승객 수 간 상관이 없다 
# H1: 지하철 호선과 월별, 승객 수 간 상관이 있다 
```

```python
lm = ols('total ~ C(line)+C(month)', data=df4).fit()
table = anova_lm(lm, typ=3) # Type 3 Anova DataFrame
print(table)
```

```python
print(f'변수 line에 대한 p_value는 {table.iloc[1,3]:.2f}, month에 대한 p_value는 {table.iloc[2,3]:.2f} 이므로 line 별 차이는 없고 월별 차이는 있다')
```

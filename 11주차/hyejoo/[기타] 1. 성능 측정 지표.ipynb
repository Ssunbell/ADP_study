## 1-6. MPE (Mean Percentage Error)

$$ MPE = {1 \over n} \sum_{i = 1}^n ({Y_i - \hat{Y_i} \over Y_i}) \times 100 \% $$

```python
# MPE       .
import pandas as pd

def MPE(y_test, y_pred) :
    return np.mean((y_test - y_pred) / y_test) * 100

def MPE2(data) :
    # Input DataFrame  
    data['Percentage Error'] = (data['Real Value'] - data['Predicted Value']) / data['Real Value'] * 100
    Mean_Percentage_Error = data['Percentage Error'].mean()
    
    return Mean_Percentage_Error
```

```python
## 나이 999살, SkinThickness 100 인 outlier 존재 -> 이상치 제거필요
def outliers_iqr(dt, col):
    q1, q3 = np.percentile(df[col], [25,75])
    iqr = q3-q1
    lower_whis = q1 - (iqr*1.5)
    upper_whis = q3 + (iqr*1.5)
    outliers = dt[(dt[col] > upper_whis) | (df[col] < lower_whis)]
    return outliers[[col]]

outliers_iqr(df, 'SkinThickness')
```

```python
## 이상치 제거할 건수 보기
for i, feature in enumerate(df.columns[:-1]):
        print(feature, outliers_iqr(df, feature).count())
```

```python
## 이상치 제거
for i, feature in enumerate(df.columns[:-1]):
    df = df.drop(index = outliers_iqr(df, feature).index)
```

```python
## 제거 후 boxplot
fig, axs = plt.subplots(figsize=(20,8), ncols=4, nrows=2)
lm_features = df.columns[:-1]

for i, feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4
    sns.boxplot(x="Outcome", 
            y=feature, 
            data=df, ax=axs[row][col])
```

### 1.2 클래스 불균형을 처리하시오.

#### 1.2.1. 업 샘플링 과정 설명하고 결과 작성

클래스불균형은 소수의 데이터 target의 중요도를 낮게 판단하여 모델링의 정확도를 낮출 가능성이 있어 데이터 분석 시 고려해야 함
이러한 불균형 해결방법으로 업샘플링, 언더샘플링 있음
업샘플링 : 소수 레이블 데이터를 다수 레이블 데이터 수만큼 증가시키는 방법. 데이터 손실없어 언더보다 성능유리하여 주로 사용

```python
# !pip install imbalanced-learn
```

```python
## overampling
x = df.drop('Outcome', axis=1)
y = df['Outcome']

from imblearn.over_sampling import RandomOverSampler
from collections import Counter
oversample = RandomOverSampler(sampling_strategy='minority') # minority = 1:1 비율 / 0.5 = 1 : 0.5 비율
x_over, y_over = oversample.fit_resample(x,y)
print(Counter(y_over))


from imblearn.over_sampling import SMOTE
smote_sample = SMOTE(sampling_strategy='minority')
x_sm, y_sm = smote_sample.fit_resample(x,y)
print(Counter(y_sm))
```

```python
df['Outcome'].value_counts()
```

#### 1.2.2 언더 샘플링 과정 설명하고 결과 작성

언더샘플링 : 다수의 레이블가진 데이터를 샘플링하여 소수 레이블 수 수준으로 감소시키는 방법, 데이터수가 줄어들어 학습성능 떨어뜨릴 수 있음

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy='majority')
x_under, y_under = undersample.fit_resample(x,y)
print(Counter(y_under))
```

#### 1.2.3 둘 중 선택하고 이유 설명

오버샘플링 중 smote 방식 데이터수 확보하여 모델 정확도 높이기 위함

### 1.3. 모델링 하시오

#### 1.3.1. 최소 3개 이상 알고리즘 제시하고 정확도 측면의 모델 1개와 속도 측면의 모델 1개를 꼭 구현(총 2개 이상)

```python
## target 변수 범주형 : classification - 
# 정확도 : light GBM
# 속도측면 : logistic regression

## 데이터셋 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_sm, y_sm, test_size=0.2, random_state=42)

## 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # 객체생성
scaler.fit(X_train) # fitting
X_train = scaler.transform(X_train) # 데이터변환 -> transform 함수는 numpy nddary로 반환(dataframe 변환)
X_test = scaler.transform(X_test)

## light gbm
## smote 방식 데이터셋으로 함

## 모델링 - 기본모델
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
import time
from sklearn.metrics import accuracy_score

# 위의 모델링 진행
dt_clf = DecisionTreeClassifier(random_state=42)
rf_clf = RandomForestClassifier(random_state=42)
lr_clf = LogisticRegression(solver='liblinear', random_state=42)

# decisiontree 학습/예측/평가
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DecisionTree 정확도:{0:.4f}'.format(accuracy_score(y_test,dt_pred)))

# randomforest 학습/예측/평가
# 랜덤포레스트 (배깅 bagging = bootstrap aggregating의 줄임말), subset 데이터 데이터중첩되어 만들어짐
# 하이퍼파라미터
# n_estimator = 3 : 전체데이터를 3개의 결정트리기반 학습하는것(서브세트 3개) default=10, 많이설정하면 좋은성능이나 무조건 향상은아님.시간오래걸림
# max_feature = 결정트리(none), 랜덤포레스트(auto=sqrt) 트리분할하는 피쳐 참조시 전체피쳐아니라 sqrt(개수) 만큼 참조
# max_depth, min_samples_leaf, min_samples_split : 결정트리에서 과적합 개선하기 위해 사용되는 파라미터가 랜덤포레스트도 동일 적용

start_time = time.time()
rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RandomForest 정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))
print('Light GBM수행시간:{0:.1f}초'.format(time.time()-start_time))

# logistic
lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('logistic 정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
print('Light GBM수행시간:{0:.1f}초'.format(time.time()-start_time))
```

```python
from lightgbm import LGBMClassifier

start_time = time.time()
lgbm_wrapper = LGBMClassifier(n_estimators=300, learning_rate=0.1, random_state=42)
lgbm_wrapper.fit(X_train, y_train, eval_metric='logloss', verbose=True)

lgbm_preds = lgbm_wrapper.predict(X_test)
print('Light GBM 정확도:{0:.4f}'.format(accuracy_score(y_test, lgbm_preds)))
print('Light GBM수행시간:{0:.1f}초'.format(time.time()-start_time))
```

#### 1.3.2. 모델 비교하고 결과 설명

트리모델 기반으로 앙상블을 모델링을 하는 경우 기존의 트리모델의 과적합 문제를 개선하여 학습데이터 서브셋 증가시킬수록
예측 정확도가 높아짐. 하지만 하이퍼파라미터(학습데이터 서브셋, 학습률 등)을 조절해줘야하므로 학습시간이 오래걸림

따라서 질병을 예측하는 모델처럼 연구목적이 정확도라면 light gbm 모델을 사용하고
학습시간을 고려한다면 rogistic regression 모델 사용

#### 1.3.3. 속도 개선을 위한 차원 축소 설명하고 수행, 예측 성능과 속도 비교하고 결과 작성

유용하지 않거나 상관관계가 높은 유사변수가 존재하는 경우 변수가 많아짐에 따라 정확한 모델링이 될 수 없을 가능성 높음
따라서 상관관계가 높은 유사변수들을 기준으로 차원을 축소하는 주성분 분석방법이 있음

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=7)
pca.fit(X_train, y_train)
print('고유값:', pca.singular_values_)
print('분산설명력:', pca.explained_variance_ratio_)
## 고유값 5개 사용 -> 80%
```

```python
plt.plot(pca.explained_variance_ratio_, 'o-')
plt.show()
```

```python
## 5개 차원으로 변수 축소하여 fitting
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train) # 데이터 변환
X_test_pca = pca.transform(X_test)

from lightgbm import LGBMClassifier

start_time = time.time()
lgbm_wrapper_pca = LGBMClassifier(n_estimators=300, learning_rate=0.1, random_state=42)
lgbm_wrapper_pca.fit(X_train_pca, y_train, eval_metric='logloss', verbose=True)

lgbm_preds_pca = lgbm_wrapper_pca.predict(X_test_pca) # 테스트는 X_test로!
print('Light GBM 정확도:{0:.4f}'.format(accuracy_score(y_test, lgbm_preds_pca)))
print('Light GBM수행시간:{0:.1f}초'.format(time.time()-start_time))
```

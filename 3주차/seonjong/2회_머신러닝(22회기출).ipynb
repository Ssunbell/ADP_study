## 1. 기계학습(50점)

### 1.1 데이터 탐색

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

df1 = pd.read_csv('../data/diabetes_for_test.csv')
df1
```

```python
df1.info() # not null
```

```python
pd.set_option('display.max_rows', None)
df1.groupby('Outcome').describe().T
```

```python
pd.set_option('display.max_rows', 10)

target = 'Outcome'
features = list(df1.columns)
features.remove(target)
r,c = 2, 4
fig, axes = plt.subplots(r, c, figsize=(10, 7))
for i, column in enumerate(features):
    ax_r, ax_c = int(i//c), int(i%c)
    sns.boxplot(y=df1[column], ax = axes[ax_r][ax_c], x=df1['Outcome'])
    axes[ax_r][ax_c].set_title(column)

plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df1.corr(), annot=True)
```

(2) 이상치를 식별하고 처리하시오. 

```python
df = df1.copy()
obnormal_idx = df[(df['Age'] > 100) | (df['BMI'] < 1) | (df['BloodPressure'] < 1) | (df['Glucose'] < 1)].index

for index in obnormal_idx:
    df.drop(index, axis=0, inplace=True)
df = df.reset_index().drop('index', axis=1)
df
```

#### 1.1.2. 앞선 두 단계에서 얻은 향후 분석시 고려사항 작성

```python
sns.heatmap(df.corr(), annot=True)
```

이상치를 제거하고 난 후 상관관계를 보면 전체적으로 증가하는 모습을 보인다.

특히, BloodPressure 및 Age에서 상관관계가 유의미하게 상승한 것을 보아

이상치에 민감한 데이터이므로 데이터 엔지니어링이 중요하다고 판단

### 1.2 클래스 불균형을 처리하시오.

#### 1.2.1. 업 샘플링 과정 설명하고 결과 작성

업 샘플링은 데이터의 수가 적은 클래스를 증가시켜 데이터 불균형을 해결하는 방법을 의미하며, 원본 데이터의 수가 적은 클래스를 복제하거나, 복제하는 동시에 약간의 노이즈를 추가하거나, 새로운 샘플을 생성하는 방식으로 클래스의 수를 늘린다.

```python
l_nums = df['Outcome'].value_counts()
```

```python
# Random Sampling
# 단순히 복사
index_list = df[df['Outcome'] == 1].index.tolist()

iloc_list = np.random.choice((index_list), size = l_nums[0] - l_nums[1])
augmentation_df = df.iloc[iloc_list]
augmentation_df = pd.concat([df, augmentation_df])
augmentation_df['Outcome'].value_counts()
```

```python
# SMOTE(KNN)
from imblearn.over_sampling import SMOTE

smote_sample = SMOTE(sampling_strategy='minority')
X, Y = df[features], df[target]
x_sm, y_sm = smote_sample.fit_resample(X, Y)
y_sm.value_counts()
```

#### 1.2.2 언더 샘플링 과정 설명하고 결과 작성

데이터의 수가 적은 클래스의 갯수에 맞게 데이터의 수가 큰 클래스의 갯수를 감소시키는 기법으로 데이터 불균형을 해결할 수 있지만 너무 많은 데이터를 감소시킬 경우 학습에 충분한 데이터 확보가 어려워 성능이 떨어질 수 있다.

```python
from imblearn.under_sampling import RandomUnderSampler

undersample = RandomUnderSampler()
x, y = undersample.fit_resample(X, Y)
y.value_counts()
```

#### 1.2.3 둘 중 선택하고 이유 설명

언더샘플링을 하게 될 경우 약 250개의 데이터만 가지게 되므로 데이터의 갯수가 너무 적어진다.

### 1.3. 모델링 하시오

#### 1.3.1. 최소 3개 이상 알고리즘 제시하고 정확도 측면의 모델 1개와 속도 측면의 모델 1개를 꼭 구현(총 2개 이상)

```python
from sklearn.model_selection import train_test_split
from xgboost.sklearn import XGBClassifier
import lightgbm as lgb
import sklearn.svm as svm
from sklearn.metrics import accuracy_score, f1_score
import time

train_x, test_x, train_y, test_y = train_test_split(df[features], df[target], test_size=0.1, random_state=42)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
start = time.time()
clf = XGBClassifier(random_state=42).fit(train_x,train_y)
preds = clf.predict(test_x)
end = time.time()
print('accuracy',accuracy_score(preds, test_y))
print('f1_score', f1_score(preds, test_y))
print('걸리는 시간', end-start)

# clf.score(preds, test_y)
```

```python
from sklearn.linear_model import LogisticRegression

start = time.time()
logistic = LogisticRegression(random_state=42)
logistic.fit(train_x, train_y)
preds = logistic.predict(test_x)

end = time.time()
print('accuracy',accuracy_score(preds, test_y))
print('f1_score', f1_score(preds, test_y))
print('걸리는 시간', end-start)
```

#### 1.3.2. 모델 비교하고 결과 설명

```python
start = time.time()
svm_clf =svm.SVC(kernel = 'linear', random_state=42).fit(train_x, train_y)
preds = svm_clf.predict(test_x)

end = time.time()
print('accuracy',accuracy_score(preds, test_y))
print('f1_score', f1_score(preds, test_y))
print('걸리는 시간', end-start)
```

#### 1.3.3. 속도 개선을 위한 차원 축소 설명하고 수행, 예측 성능과 속도 비교하고 결과 작성

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

tmp = StandardScaler().fit_transform(df[features])
pca = PCA(n_components=len(features))
pca.fit(tmp)

features_p = [(ratio, feature) for feature, ratio in zip(features, pca.explained_variance_ratio_)]
for r, f in features_p:
    print(r, f)
    
plt.plot(pca.explained_variance_ratio_, 'o-')
plt.show()
```

```python
features_p = [row[1] for row in features_p][:6]
tmp = StandardScaler().fit_transform(df[features_p])
pca = PCA(n_components=6)
pca.fit(tmp)
new_features = pd.DataFrame(tmp, columns=[features_p])
new_features
```

```python
train_x2, test_x2, train_y2, test_y2 = train_test_split(new_features, df['Outcome'], test_size=0.1, random_state=42)

start = time.time()
clf = XGBClassifier(random_state=42).fit(train_x2,train_y2)
preds = clf.predict(test_x2)
end = time.time()
print('XGBoost accuracy',accuracy_score(preds, test_y2))
print('XGBoost f1_score', f1_score(preds, test_y2))
print('XGBoost 걸리는 시간', end-start)

start = time.time()
logistic = LogisticRegression(random_state=42)
logistic.fit(train_x2, train_y2)
preds = logistic.predict(test_x2)

end = time.time()
print('logistic accuracy',accuracy_score(preds, test_y2))
print('logistic f1_score', f1_score(preds, test_y2))
print('logistic 걸리는 시간', end-start)

start = time.time()
svm_clf =svm.SVC(kernel = 'linear', random_state=42).fit(train_x2, train_y2)
preds = svm_clf.predict(test_x2)

end = time.time()
print('SVM accuracy',accuracy_score(preds, test_y2))
print('SVM f1_score', f1_score(preds, test_y2))
print('SVM 걸리는 시간', end-start)
```
